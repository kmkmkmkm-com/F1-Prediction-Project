```python
import fastf1
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import os

fastf1.Cache.enable_cache('f1_cache')

race_names = [
    'Australian Grand Prix', 'Chinese Grand Prix', 'Japanese Grand Prix',
    'Bahrain Grand Prix', 'Saudi Arabian Grand Prix', 'Miami Grand Prix',
    'Emilia Romagna Grand Prix', 'Monaco Grand Prix', 'Spanish Grand Prix',
    'Canadian Grand Prix', 'Austrian Grand Prix', 'British Grand Prix',
    'Belgian Grand Prix', 'Hungarian Grand Prix', 'Dutch Grand Prix',
    'Italian Grand Prix', 'Azerbaijan Grand Prix', 'Singapore Grand Prix',
    'United States Grand Prix', 'Mexico City Grand Prix', 'São Paulo Grand Prix',
    'Las Vegas Grand Prix', 'Qatar Grand Prix', 'Abu Dhabi Grand Prix'
]

all_sessions = {}
for race in race_names:
    session = fastf1.get_session(2025, race, 'R')
    session.load(telemetry=False, weather=False, messages=False)
    all_sessions[race] = session
    print(f"Loaded: {race}")

print(f"\nTotal: {len(all_sessions)} races")
```

```python
driver_colors = {
    'VER': '#3671C6', 'NOR': '#FF8000', 'LEC': '#E8002D',
    'PIA': '#FF8000', 'SAI': '#E8002D', 'HAM': '#27F4D2',
    'RUS': '#27F4D2', 'ALO': '#358C75', 'PER': '#3671C6',
    'STR': '#358C75', 'ANT': '#27F4D2', 'ALB': '#64C4FF'
}

pace_data = []
for race, session in all_sessions.items():
    laps = session.laps
    laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()
    clean = laps[laps['IsAccurate'] == True].copy()
    median_pace = clean.groupby('Driver')['LapTimeSeconds'].median()
    fastest = median_pace.min()
    for driver, pace in median_pace.items():
        pace_data.append({
            'Race': race,
            'Driver': driver,
            'MedianLapTime': pace,
            'GapToFastest': pace - fastest
        })

pace_df = pd.DataFrame(pace_data)
print(pace_df.shape)
```

```python
wins_data = []
for race, session in all_sessions.items():
    results = session.results.sort_values('Position')
    podium = results.head(3)[['Abbreviation', 'Position']].copy()
    podium['Race'] = race
    wins_data.append(podium)

wins_df = pd.concat(wins_data)

# Count wins per driver
win_counts = wins_df[wins_df['Position'] == 1]['Abbreviation'].value_counts()
podium_counts = wins_df['Abbreviation'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(16, 6), facecolor='#1a1a2e')
for ax in axes:
    ax.set_facecolor('#1a1a2e')

# Wins
colors_wins = [driver_colors.get(d, '#888888') for d in win_counts.index]
axes[0].barh(win_counts.index, win_counts.values, color=colors_wins)
axes[0].set_title('Wins (2025 Season)', color='white', fontsize=13)
axes[0].set_xlabel('Count', color='white')
axes[0].tick_params(colors='white')
axes[0].invert_yaxis()

# Podiums
colors_pods = [driver_colors.get(d, '#888888') for d in podium_counts.index]
axes[1].barh(podium_counts.index, podium_counts.values, color=colors_pods)
axes[1].set_title('Podiums (2025 Season)', color='white', fontsize=13)
axes[1].set_xlabel('Count', color='white')
axes[1].tick_params(colors='white')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
circuit_gaps = []

for race, session in all_sessions.items():
    laps = session.laps.copy()
    laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()
    clean = laps[laps['IsAccurate'] == True]
    
    median_pace = clean.groupby('Driver')['LapTimeSeconds'].median().sort_values()
    
    if len(median_pace) >= 10:
        gap = median_pace.iloc[9] - median_pace.iloc[0]
        circuit_gaps.append({'Race': race, 'P1toP10Gap': gap})

gap_df = pd.DataFrame(circuit_gaps).sort_values('P1toP10Gap', ascending=False)

fig, ax = plt.subplots(figsize=(14, 7), facecolor='#1a1a2e')
ax.set_facecolor('#1a1a2e')

ax.barh(gap_df['Race'], gap_df['P1toP10Gap'], color='#FF8000')
ax.set_title('Pace Gap P1 to P10 by Circuit (2025 Season)', color='white', fontsize=13)
ax.set_xlabel('Gap (s)', color='white')
ax.tick_params(colors='white')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
top_drivers = pace_df.groupby('Driver')['GapToFastest'].mean().sort_values().head(5).index.tolist()
print(top_drivers)
```

```python
fig = go.Figure()

for driver in top_drivers:
    d_data = pace_df[pace_df['Driver'] == driver]
    fig.add_trace(go.Scatter(
        x=d_data['Race'],
        y=d_data['GapToFastest'],
        mode='lines+markers',
        name=driver,
        line=dict(color=driver_colors.get(driver, '#888888'), width=2),
        marker=dict(size=6)
    ))

fig.update_layout(
    title='Championship Pace Gap to Fastest (2025 Season)',
    xaxis_title='Race',
    yaxis_title='Gap to Fastest (s)',
    plot_bgcolor='#1a1a2e',
    paper_bgcolor='#1a1a2e',
    font=dict(color='white'),
    xaxis=dict(tickangle=45),
    hovermode='x unified'
)

fig.write_html('championship_pace_gap_2025.html')
print("Interactive chart saved as 'championship_pace_gap_2025.html'")
```
